```python
# import packages.
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
# Data Files.
X_TRAIN = "X_train.csv"
Y_TRAIN = "y_train.csv"
X_TEST = "X_test.csv"
SUB = "submit.csv"
```

```python
# Read data files to data frame.
X_train=pd.read_csv(X_TRAIN)
y_train=pd.read_csv(Y_TRAIN)
X_test=pd.read_csv(X_TEST)
```

```python
# Peek into the data.
#X_train.head()
#y_train.head()
X_test.head()
#X_train['brand'].unique()
```

```python
# Encode categorical data to one hot encoding.
train_data_final = pd.get_dummies(X_train, columns=["brand"])
train_data_final.head()
```

```python
# Merge X_train & y_train
train_data_final1=pd.merge(train_data_final, y_train, how='inner')
train_data_final1.head()
```

```python
# Drop the id feature.
train_data_final2 = train_data_final1.drop(['id'], axis=1)
train_data_final2.head()
```

```python
# Encode categorical data to one hot encoding in test data. Also drop id feature. 
test_data = pd.get_dummies(X_test, columns=["brand"])
test_data = test_data.drop(['id'], axis=1)
test_data.head()
```

```python
# To test data add the column of value with values as 0.
test_data['value'] = 0 
test_data.head()
```

```python
# Get training data & regression value separately.
XTrain = train_data_final2.iloc[:,:-1]
YTrain = train_data_final2.iloc[:,-1:]
```

```python
# Get test data separately.
XTest = test_data.iloc[:,:-1]
XTest.head()
```

```python
# Linear Regression Model
ucvLinear = linear_model.LinearRegression(normalize=True)
ucvLinear.fit(XTrain, YTrain)
ucvLinear.coef_
```

```python
# Predict for test data.
YucvTest = ucvLinear.predict(XTest)
print(YucvTest)
```

```python
# Create a data frame & keep only the id & value column.
yPred = X_test
ypred = yPred.drop(['year', 'brand', 'age', 'engine_size', 'power', 'mileage', 'prev_owners'], axis=1)
ypred['value'] = YucvTest
ypred.to_csv(SUB)
ypred.head(100)
```
